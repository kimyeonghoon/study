### request 라이브러리

```python
import requests

req = requests.get("http://yeonghoon.kim")

print(req.text) # 텍스트 정보 가져오기
```

```python
print(req.headers) # 헤더 정보 가져오기
```

```python
print(req.status_code) # 상태 코드 가져오기
```

```python
print(req.cookies) # 쿠키 값 가져오기
```

```python
print(req.content) # 컨텐츠 내용 가져오기
```
